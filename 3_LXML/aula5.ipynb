trabalhando com arquivos XML
- XML projetado para transportar dados
- Tags nao sao predefinidas, iguais o HTML (nao tem div, etc)
- As Tags sao tipo: funcionarios (elemento), nome (texto), cpf, etc

```python
from lxml import etree
```

<Teste>
    Este é o texto da tag Teste
<\Teste>

```python
elem = etree.Element("Teste")
elem.text = "Este é o texto da tag Teste"
print(etree.tostring(elem))
print(elem.tag)
```

<clientes>
    <cliente>
        <nome> Ana <\nome>
        <idade> 24 <\idade>
        <sexo> Fem <\sexo>
        <cpf> 123.456.789-00 <\cpf>
    <\cliente>
<\clientes>

```python
clientes = etree.Element("clientes")


cliente1 = etree.SubElement(clientes, "cliente")

nome1 = etree.SubElement(cliente1, "nome")
nome1.text = "Ana"

idade1 = etree.SubElement(cliente1, "idade")
idade1.text = "24"

sexo1 = etree.SubElement(cliente1, "sexo")
sexo1.text = "Fem"

cpf1= etree.SubElement(cliente1, "cpf")
cpf1.text = "123.456.789-00"

print(etree.tostring(clientes))
```

Elements sao listas
- um Element imita o comportamento de listas do python
- É possivel manipular oobjeto Element como se fosse uma lista e tambem converter em lista com list()

```python
cliente2 = etree.SubElement(clientes, "cliente2")

nome2 = etree.SubElement(cliente2, "nome")
nome1.text = "Clara"

idade2 = etree.SubElement(cliente2, "idade")
idade2.text = "20"

sexo2 = etree.SubElement(cliente2, "sexo")
sexo2.text = "Fem"

cpf2= etree.SubElement(cliente2, "cpf")
cpf2.text = "311.444.897-23"

clientes.append(cliente2)

print(etree.tostring(clientes))
```

```python
# usar como lista

print("Total de clientes: ", len(clientes))
```

```python
# criando um objeto com base no segundo elemento

cliente_dois = clientes[1]
print("Tag clientes[1]: ", cliente_dois.tag)
```

```python
for i in clientes:
    print(i.tag)
```
