Principais Caracteres especiais
# . Qualquer caractere menos o de nova linha "\n"
# ^ inicio do texto. Modo MULTILINE tb coincide imediatamente apos cada nova linha
# $ Fim do texto
# \ Permite usar caracteres especiais como se fossem comuns
# [] Qualquer caractere listado entre os conchetes [abc] ou sequencia [a-z]
# * Zero ou mais ocorrencias da expressao anterior
# + Uma ou mais ocorrencias da expressao anterior
# ? zero ou uma ocorrencia. Para quando o caractere nao é obrigatorio

Mais caracteres especiais
# \d   Digito.[0-9]
# \D   Nao digito. [^0-9]
# \s   Qualquer caract de espaçamento [\t\n\r\f\v]
# \S   Qualquer caract que nao é de espaçamento  [^ \t\n\r\f\v]
# \w   Caract alfanumerico ou sublinhado [a-z A-Z 0-9 _]
# \W   Caract que nao sejam alfanumerico ou sublinhado  [ ^ a-z A-Z 0-9 _]
# {n}  N correncias da expressao anterior
# |    'Ou' logico


FLAGS
# re.IGNORE: Nao distingue minusculas de maiusculas
# re.DOTALL: Qualquer caractere, incluindo uma nova linha

```python
texto = ("\nOlá, esta é uma aula de python. Esta aula é sobre Expressões Regulares. Espero que goste desta aula.")
padrao = "[aáoõ]"
resultado = re.findall(padrao, texto)
print(resultado)

padrao2 = "[A-Z]"
resultado2 = re.findall(padrao2, texto)
print(resultado2)
```

```python
texto = ("\nOlá, esta é uma aula de python. Esta aula é sobre Expressões Regulares. Espero que goste desta aula.")
padrao = "(au)*"
resultado = re.findall(padrao, texto)
print(resultado)


padrao2 = "(au)+"
resultado2 = re.findall(padrao2, texto)
print(resultado2)
```

```python
texto = ("\nOlá, esta é uma aula de python. Esta aula é sobre Expressões Regulares. Espero que goste desta aula.")
padrao = "(au)*"
resultado = re.findall(padrao, texto)
print(resultado)


padrao2 = "(au)+"
resultado2 = re.findall(padrao2, texto)
print(resultado2)
```

```python
texto = ("\nOlá 0975, esta é a aula 6 de python. Esta aula 91 é sobre Expressões Regulares. Espero que goste desta aula 01.")
padrao = "\d"
resultado = re.findall(padrao, texto)
print(resultado)


padrao2 = "\D"
resultado2 = re.findall(padrao2, texto)
print(resultado2)

# que possuem 2 digitos
padrao3 = "\d{2}"
resultado3 = re.findall(padrao3, texto)
print(resultado3)
```

```python
texto = "Na segunda-feira eu nao vou trabalhar. Na terça-feira nao vou descansar. Na quarta preciso recuperar. Na quinta-feira eu acordo meio dia. Ja na sexta viajo."

padrao = re.compile(r"(segunda|terça|quarta|quinta|sexta)(-feira)*")
resultado = re.findall(padrao, texto)
print(resultado)

padrao2 = re.compile(r"(segunda|terça|quarta|quinta|sexta).feira")
resultado2 = re.findall(padrao2, texto)
print(resultado2)
```

CPF
XXX.XXX.XXX-XX
XXXXXXXXXXX

[0-9]{3} [\.]? [0-9]{3} [\.]?  [0-9]{3} [\-]?  [0-9]{2}

```python
padrao = re.compile("([0-9]{3}[\.]?[0-9]{3}[\.]?[0-9]{3}[\-]?[0-9]{2})")

texto = "No processamento de dados de hoje, identificamos o cliente João com o CPF 123.456.789-00 e a Maria, cujo registro consta como 98765432111 sem pontuação. Durante a auditoria, outros documentos surgiram no relatório, como o 111.222.333-44, o antigo registro 555666777-88 que usa apenas o traço, e um caso isolado onde o sistema salvou apenas pontos como em 000.111.222.33, totalizando cinco registros válidos para o seu teste de extração."

resultado = re.findall(padrao,texto)
print(resultado)
```

E-MAIL

ana_clara1@ufv.br

[\w.-]+@ [\w.-]+

```python
padrao = re.compile(r"([\w.-]+@[\w.-]+)")

texto = "Prezados, favor encaminhar as faturas para o setor financeiro no e-mail contato.direto@empresa-servicos.com.br ou diretamente para o supervisor em ana.clara22@gmail.com. Caso ocorra algum erro no servidor interno (mail-server-01@internal.net), favor notificar o suporte técnico através do endereço suporte_tecnico@provedor.org. Não utilizem o antigo e-mail usuario@teste, pois ele está desativado, assim como o endereço incompleto @https://www.google.com/search?q=dominio.com que não deve ser capturado."

resultado = re.findall(padrao, texto)
print(resultado)

```

LOCALIZANDO DATAS

\d{2} / \d{2} / \d{4}

```python
padrao = re.compile(r"\d{2}/\d{2}/\d{4}")

texto = "A reunião de abertura do projeto foi agendada para o dia 15/03/2024, mas devido a imprevistos, foi adiada para 02/04/2024. O contrato foi assinado originalmente em 31/12/2023 às 14:30. Note que formatos como 2024-05-10 ou 10-05-2024 não devem ser capturados, assim como a data incompleta 05/05/24. O sistema deve extrair apenas as datas completas com quatro dígitos no ano, como 25/12/2025."

resultado = re.findall(padrao, texto)
print(resultado)
```
