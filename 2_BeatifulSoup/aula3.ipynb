```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from urllib.error import HTTPError, URLError
import re
```

```python
try:
    html = urlopen("https://ccp.ufv.br/historico/")
except HTTPError as erro:
    print(erro)
except URLError as erro:
    print(erro)
```

```python
# transformar o conteudo html em um objeto beautifulSoup
bsObj = BeautifulSoup(html.read(), "html.parser")
```

```python
print(bsObj.title)
print(bsObj.h1)
```

```python
for link in bsObj.find_all('a'):
    print(link)
```

```python
for link in bsObj.find_all('a'):
    print(link.get('href'))
```

```python
teste = bsObj.find_all("li", {"class": "list-item"})

for i in teste:
    link = i.find("a")
    if link:
        print(link.get_text().strip())
```

```python
try:
    response = urlopen("https://ge.globo.com/")
    html= response.read() 
    bsObj = BeautifulSoup(html, "html.parser")
except HTTPError as erro:
    print(erro)
except URLError as erro:
    print(erro)
```

```python
serie_a_g6 = ('.athetico.|.atlético.|.bahia.|.botafogo.|.bragantino.|.cruzeiro.|flamengo')
```

```python
for link in bsObj.findAll("a", href=re.compile(serie_a_g6)):
    if "href" in link.attrs:
        nova_pag = link.attrs['href']
        print(nova_pag)
```
